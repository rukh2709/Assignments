Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Answer 1: Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in Python. 

Matplotlib is used to create visualizations and graphs to represent data in a visually appealing way. 

Following are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot is a basic chart that displays data points connected by straight lines. It is used to show trends and patterns over time.

Scatter Plot: A scatter plot is used to show the relationship between two variables. It shows how much one variable is affected by another.

Bar Plot: A bar plot is used to compare different categories of data. It is used to show the distribution of data among categories.

Histogram: A histogram is used to show the distribution of numerical data. It divides data into intervals and shows how many values fall into each interval.

Pie Chart: A pie chart is used to show the proportion of data in different categories. It is a circular graph that shows the relationship between the parts and the whole.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Answer 2: A scatter plot is a type of plot or chart that displays the relationship between two variables. 

```python
import matplotlib.pyplot as plt
import numpy as np
a=np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
sizes=50*np.random.rand(50)
plt.scatter(x, y, c='b', s= sizes, alpha=0.4, marker='*')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer 3: The subplot() function in Matplotlib is used to create multiple subplots or plots within a single figure. It allows us to create and arrange multiple plots in a grid-like layout, each with their own axes and plot settings.

```python
import matplotlib.pyplot as plt
import numpy as np
#For line 1:
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
```

```python
#For line 2: 
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
```

```python
#For line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
```

```python
#For line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(5, 4))

# Plot each line on a different subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Set the title 
fig.suptitle('Four Line Plots')
#adjust the spacing between subplots
fig.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Answer 4: A bar plot (also known as a bar chart) is a graphical representation of data that uses rectangular bars to show the values of different categories. In a bar plot, the height or length of each bar corresponds to the numerical value or frequency of the category it represents.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit earned")
plt.title("Different companies with its profit earned")
plt.show()
```
