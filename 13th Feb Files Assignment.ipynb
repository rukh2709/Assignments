Q1. Explain why we have to use the Exception class while creating a Custom Exception.

Answer 1: To create a custom exception class, we define a class that inherits from the built-in Exception class or one of its subclasses such as ValueError class.
The following example defines a CustomException class that inherits from the Exception class:

class CustomException(Exception):
    """ my custom exception class """

To raise the CustomException, we use the raise statement. For example, the following uses the raise statement to raise the CustomException:

```python
class CustomException(Exception):
    """ my custom exception class """
try:
    raise CustomException('This is my custom exception')
except CustomException as e:
    print(e)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Answer 3: The ArithmeticError Exception is the base class for all errors associated with arithmetic operation.

ArithmeticError
(1) ZeroDivisionError
(2) OverflowError
(3) FloatingPointError

```python
#Example - Handling ZeroDivisionError
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
#Example - Handling OverflowError
j = 2.0
try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
#Example - Handling FloatingPointError
a = 1.1
b = 2.2
c = 3.3

try:
    result = c - (a + b)
except FloatingPointError:
    print("A floating-point error occurred!")
else:
    print("The result is:", result)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
Answer 4: To handle both IndexError and KeyError exception classes we use LookupError exception class.
```

```python
#Example - Handling IndexError exception
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```

```python
#Example - Handling KeyError exception
student = {'name': 'Nipun Mishra',
                'age': 26,
                'language taken': 'Python'}
user_input = input('What do you want to learn about student==> ')

try:
    print(f'{user_input} is {student[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

Q5. Explain ImportError. What is ModuleNotFoundError?

Answer 5: In Python, ImportError occurs when the Python program tries to import module which does not exist in the private table. This exception can be avoided using exception handling using try and except blocks. 
ModuleNotFoundError error occurs when we are trying to access or use a module that cannot be found.
