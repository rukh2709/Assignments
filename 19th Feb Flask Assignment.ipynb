This means that there are a lot of resources available for developers who are using Flask, including documentation, tutorials, and code examples.

Extensibility: Flask is highly extensible, allowing developers to add new functionality through its modular design. This means that developers can easily customize and extend Flask to meet their specific needs.

RESTful API Support: Flask has built-in support for building RESTful APIs, which makes it a popular choice for building web services and APIs.

Q3. What is App routing in Flask? Why do we use app routes?

Answer 3: In Flask, app routing is the process of mapping URLs to Python functions that handle those requests. App routes define the URLs that can be accessed by users and the functions that should be called when a request is made to a specific URL.

We use app routes in Flask to:

Define URLs: App routes are used to define the URLs that can be accessed by users, which makes it easy for users to navigate through different parts of the application.

Handle Requests: App routes are used to handle incoming requests from users and provide appropriate responses based on the requested URL.

Create Endpoints: App routes allow developers to create endpoints in the application, which are the points where users can interact with the application and perform different actions.

Organize Code: App routes allow developers to organize the code for the application based on the different URLs and endpoints that it handles.

Here's an example of a basic app route in Flask:

from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'
    
In this example, the @app.route('/') decorator maps the URL / to the index function, which returns the string "Hello, World!" when a user accesses the root URL of the application. This is a very basic example, but in practice, app routing can become much more complex and is a critical part of building a Flask application.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

Answer 5: In Flask, the url_for() function is used for URL building.

```python
# Python code to demonstrate the working of the url_for() function.
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def profile(username):
    return f'{username}\'s profile'

with app.test_request_context():
    print(url_for('index'))
    print(url_for('profile', username='Sumaiya'))    

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

In this example, we import the Flask class and the url_for function from the flask module. We then create a new Flask application instance and define two routes: the home page and a dynamic route that accepts a user name.

In the if __name__ == '__main__' block, we use the test_request_context() function to create a test request context, which allows us to use the url_for() function without running the server. We then call the url_for() function for each of the two routes and print the result.
