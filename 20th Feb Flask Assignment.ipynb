Q1. Explain GET and POST methods.

Answer 1: 
The GET method is used to retrieve data from a web server. When you click on a link or enter a URL in your web browser, you are using the GET method to retrieve the HTML code that makes up the web page. The data is sent as part of the URL itself, and is visible in the address bar of your browser. GET requests are generally used for retrieving data that is publicly available, such as a web page or an image.

The POST method is used to submit data to a web server. When you fill out a form on a website and click "submit", you are using the POST method to send the data you entered to the web server for processing. The data is sent in the request body, rather than as part of the URL, which means that it is not visible in the address bar. POST requests are generally used for submitting data that is not publicly available, such as login credentials or sensitive information.

Q2. Why is request used in Flask?

Answer 2: In Flask, the request module is used to access the information that a client (e.g., a web browser) sends to a web server during an HTTP request. This information includes data submitted in a form, headers sent by the client, and more.

Q3. Why is redirect() used in Flask?

Answer 3: In Flask, the redirect() function is used to redirect a user to a different URL. It's a way of telling the client (e.g., a web browser) to go to a new location, and is commonly used in web applications to handle certain types of user interactions.

Q4. What are templates in Flask? Why is the render_template() function used?

Answer 4: In Flask, templates are used to separate the presentation logic (how the data is displayed) from the business logic (how the data is processed).
The render_template() function is used to render a template in Flask and to generate HTML files from templates. By using templates, developers can create more maintainable and reusable code, and provide a better user experience for their web applications.

Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
#Answer 5:
#Created a simple API
from flask import Flask, request, jsonify

app=Flask(__name__)

@app.route('/postman_data', methods=['POST'])
def age_group():
    if request.method=='POST':        
        age=int(request.json['age'])  

        if age<=1:            
            result= 'You are an infant'

        elif age>1 and age<=13:
            result= 'You are a child'
       
        elif age>13 and age<=19:
            result= 'You are a teenager'
        
        elif age>19 and age<=60:
            result= 'You are an adult'

        elif age>6:
            result= 'You are a senior citizen'

        return jsonify(result)
if __name__=="__main__":
    app.run(host="0.0.0.0")
```
